```python
# прочитаем и занесем файл в переменную data
data = pd.read_csv('games.csv') 
```

При изучении общей информации данных можно заметить наличие в таблице пустых значений в столбцах "Name", "Year_of_Release", "Genre", "Critic_Score", "User_Score", неправильный тип данных у столбцов "Year_of_Release" и "User_Score", неправильное название столбцов (присустствует верхний регистр).
Пустые значения в "Critic_Score" и "User_Score" могут говорить об отсутствии рейтинга у данных игр.

### Предобработка данных

#### Приведем столбцы к нижнему регистру

```python
data.columns = data.columns.str.lower() 
```

#### Обработка пропущенных значений

```python
data.info()
```

Удалим строки, где не указано название игры и год издания. Их не так много, поэтому это не сильно повлияет на данные

```python
data = data.dropna(subset=['name', 'year_of_release']) 
```

```python
data.info()
```

```python
data['rating'].unique()
```

Рейтинг 'E' («Everyone») означает то же самое, что и 'K-A' ("Kids to Adults"), поэтому их можно объединить, заменив все на 'E'

```python
# произведем замену 'K-A' на 'E'
data['rating'] = data['rating'].replace('K-A', 'E') 
```

```python
# оставшиеся пустые значения заменим 'no info'
data['rating'] = data['rating'].fillna('no info')
```

```python
data['rating'].value_counts()
```

```python
# просмотрим уникальные значения столбца 'user_score'
data['user_score'].unique() 
```

Среди уникальных значений, помимо пустых (nan), найдено не числовое значение tbd, которое дословно обозначает "to be determined", то есть рейтинг игры еще не сформирован. Проверим, соответствует ли данный рейтинг пользователей исключительно новым играм.

```python
np.sort(data[data['user_score'] == 'tbd']['year_of_release'].unique())
```

Как видно из таблицы, значение "tbd" встречается в играх как старых годов выпуска, так и новых(в соответствии с задачей). Проверим, можно ли удалить данные значения, посчитав их количество в данных.

```python
(data['user_score'].value_counts()/data.shape[0]).sort_values(ascending=False)
```

Значение tbd занимает более 14% в общем объеме данных, следовательно просто удалить его нельзя. Заменить на 0 нельзя, так как в списке есть игры с таким рейтингом, на "no info" - нельзя, так как необходимо привести этот столбец к типу данных float. Остается заменить nan и tbd на любое другое число вне диапазона от 0 до 10. Возьмем для замены число 404

```python
data['user_score'] = data['user_score'].fillna(-404)
data['user_score'] = data['user_score'].replace('tbd', -404)
```

```python
# проверим, осуществилась ли замена
(data['user_score'].value_counts()/data.shape[0]).sort_values(ascending=False)
```

```python
# просмотрим уникальные значения столбца 'critic_score'
data['critic_score'].unique() 
```

Заменим пустые значения также на -404

```python
data['critic_score'] = data['critic_score'].fillna(-404)
```

#### Замена типов данных

Неправильный тип данных указан у двух столбцов: year_of_release и user_score. year_of_release достаточно будет привести к значению int, так как не интересует месяц, день, час, минута и секунда выпуска. user_score необходимо привести к числовому типу с плавающей точкой.

```python
# заменим тип данных в столбцах
data['user_score'] = data['user_score'].astype('float') 
data['year_of_release'] = data['year_of_release'].astype('int')
```

Все данные приведены к правильным типам

#### Подсчет продаж по странам

```python
# создадим столбец с суммой продаж по всем странам
data['sum_sales'] = data[['na_sales', 'eu_sales', 'jp_sales']].sum(axis=1)
```

#### Влияние на продажи отзывов пользователей и критиков на примере PS4

```python
# cоздадим отдельный датафрейм по PS4 
# избавимся от замененных пустых значений в столбцах critic_score и user_score
data_ps4 = data[(data['platform'] == 'PS4') & 
                (data['critic_score'] > 0) & 
                (data['user_score'] > 0)] 
```

```python
data_ps4.head()
```

```python
# cоздадим функцию для построения графиков и корреляции продаж и отзывов пользователей и критиков на примере различных платформ 
def platform_scattergraph(df):
    sns.heatmap(df[['sum_sales', 'critic_score', 'user_score']].corr(), annot=True)
    plt.title('Корреляция продаж и оценок зрителей и критиков платформы {}'.format(df.platform.unique()))
    for i, x in enumerate(["critic_score", "user_score"]):
        df.plot(kind='scatter', 
                x=x, 
                y='sum_sales', 
                alpha=0.3,  
                ylabel='Суммарные продажи, млн копий')
        if x == "critic_score":
            plt.xlabel ('Оценка критиков')
            plt.title(f'Влияние на продажи отзывов критиков на примере {df.platform.unique()}')
            print(f'Корреляция продаж и оценки критиков игр на примере {df.platform.unique()}:', 
                  df[x].corr(df["sum_sales"]))
        else:
            plt.xlabel ('Оценка пользователей')
            plt.title(f'Влияние на продажи отзывов пользователей на примере {df.platform.unique()}')
            print(f'Корреляция продаж и оценки пользователей игр на примере {df.platform.unique()}:', 
                  df[x].corr(df["sum_sales"]))
            
```

```python
platform_scattergraph(data_ps4) #Найдем влияние на примере PS4
```

В результате построения диаграмм рассеяния и подсчета корреляции между продажами и оценкой критиков и пользователей было выяснено, что между оценкой критиков и продажами существует слабая корреляция в примерно 40%. Между оценкой пользователей и продажи корреляция отрицательная и равна нулю, то есть эти значения никак не зависят друг от друга. Несмотря на низкий рейтинг среди пользователей продажи могут быть высокими. Возможно это связано с тем, что консоль вышла относително недавно (ориентируясь на 2016 год), чем и был вызван ажиотаж выпускаемых на нее игр. Раз между оценками критиков и продажами существует небольшая связь, значит пользователи обращают на это внимание, но редко

### Проверка гипотез

Для проверки гипотез о равенстве средних двух генеральных совокупностей воспользуемся специальным методом: scipy.stats.ttest_ind (array1, array2, equal_var)

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую гипотезу и альтернативную, которая должна противоречить нулевой:

    H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые. 
    H1 - Средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
# создадим функцию для сравнения гипотез
def h_compare (value, criterion):
    
    # зададим критерий alpha
    alpha = .05 
    
    # создадим выборки
    df1 = data[(data[criterion] == value[0]) & (data['user_score'] >= 0)]['user_score']
    df2 = data[(data[criterion] == value[1]) & (data['user_score'] >= 0)]['user_score']
    
    # построим гистограммы выборок
    fig, axs = plt.subplots(2)
    fig.suptitle(f'Сравнение распределения выборок по {value[0]} и {value[1]}, отражающих пользовательский рейтинг')
    axs[0].hist(df1)
    axs[1].hist(df2)
    
    # превратим выборки в массивы для проверки гипотез
    df1 = np.array(df1) 
    df2 = np.array(df2) 
    
    # выведем дисперсии выборок
    print(f'Дисперсия выборок {value[0]} и {value[1]}: ', '\n', np.var(df1, ddof=1), '\n', np.var(df2, ddof=1))
    results = st.ttest_ind(df1, df2)
    
    # найдем значение p-value
    print('p-значение:', results.pvalue)
    
    # сравним p-value с alpha
    if results.pvalue < alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
criterion = ['XOne', 'PC']
h_compare(criterion, 'platform')
```

Гистограммы обеих выборок указывают на их отрицательную скошенность.

Нельзя отвергать нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC.
