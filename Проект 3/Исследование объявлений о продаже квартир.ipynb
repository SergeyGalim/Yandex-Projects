### Предобработка данных

#### Обработка пропущенных значений

```python
# определим, в каких столбцах сколько пропусков
data.isna().sum() 
```

```python
# заполняем пропуски в столбце 'ceiling_height' значением моды
data['ceiling_height'] = data['ceiling_height'].fillna(statistics.mode(data['ceiling_height'])) 
```

Общее количество пропусков в значении этажей дома говорит о неполностью заполненных данных при публикации объявлений. Так как пустых данных в этом столбце мало, можно просто их удалить

```python
# удаляем строки с пустыми значениями в столбце 'floors_total'
data = data.dropna(subset=['floors_total']) 
```

Пустые значения в столбцах 'living_area' и 'kitchen_area' также говорит о незаполненности данных при объявлении. Заполнить значения можно следующим образом: найти соотношение средней жилой зоны (где она заполнена) к средней общей зоны (где заполнена площадь жилой зоны) недвижимости. Также сделать и с кухней.

```python
# находим среднее значение площади жилой зоны
liv_area_mean = data[data['living_area'].isna() == False].living_area.mean() 
```

```python
# находим среднее значение всей площади, где указана площадь жилой зоны
tot_area_mean1 = data[data['living_area'].isna() == False].total_area.mean() 
```

```python
# находим среднее значение площади кухонной зоны
kit_area_mean = data[data['kitchen_area'].isna() == False].kitchen_area.mean() 
```

```python
tot_area_mean2 = data[data['kitchen_area'].isna() == False].total_area.mean() 
```

```python
# находим соотношение жилой зоны ко всей
percent_liv_area = liv_area_mean / tot_area_mean1 
```

```python
# находим соотношение кухонной зоны ко всей
percent_kit_area = kit_area_mean / tot_area_mean2 
```

```python
print(percent_liv_area, '\n', percent_kit_area)
```

```python
# заполняем пропуски в столбце 'living_area' умножением общей зоны на полученный коэффициент
data['living_area'] = data['living_area'].fillna(data['total_area'] * percent_liv_area) 
```

```python
# заполняем пропуски в столбце 'kitchen_area' умножением общей зоны на полученный коэффициент
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] * percent_kit_area) 
```

Пропуски значений в столбце 'is_apartment' могут означать неправильный тип данных данного столбца (object)

```python
# найдем количество уникальных значений в данном столбце
data['is_apartment'].value_counts() 
```

Как видно, данный столбец должен иметь тип bool

```python
# переведем столбец в тип bool
data['is_apartment'] = data['is_apartment'].astype('bool') 
```

```python
data['is_apartment'].value_counts()
```

Пустые значения в столбце 'balcony' говорят об отсутствии балконов в недвижимости

```python
# заменим пустые значения на 0
data['balcony'] = data['balcony'].fillna(0) 
```

Общее количество пропусков в столбце 'locality_name' говорит о неполностью заполненных данных при публикации объявлений. Так как пустых данных в этом столбце мало, можно просто их удалить

```python
data = data.dropna(subset=['locality_name'])
```

Пропуски в столбце days_exposition говорят о том, что квартира не была снята с публикации с момента сбора данных. Пропуски в airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest говорят об отсутствии этих объектов рядом с недвижимостью, либо о неправильности переданных данных. Оставим пропуски без изменений

Минимальная стоимость квартиры равна 12 (в тысячах), что не может являться правдой. Отсечем квартиры стоимостью ниже 2 000 000

### Добавление новых столбцов

```python
# цена за квадратный метр
data['price_for_meter'] = data['last_price'] / data['total_area']
```

```python
# день недели публикации объявления
data['weekday'] = data['first_day_exposition'].dt.weekday
```

```python
# месяц публикации объявления
data['month'] = data['first_day_exposition'].dt.month
```

```python
# год публикации объявления
data['year'] = data['first_day_exposition'].dt.year
```

```python
# тип этажа квартиры
def floor_type(df): 
    if df['floor'] == 1:
        return 'первый'
    if df['floor'] == df['floors_total']:
        return 'последний'
    return 'другой'

# создадим новый столбец с типом этажа, применив функцию в нашей таблице
data['floor_type'] = data.apply(floor_type, axis = 1) 
```

```python
# расстояние до центра города в километрах
data['cityCenters_nearest_km'] = round(data['cityCenters_nearest'] / 1000)
```

#### Средняя цена каждого километра до центра в Санкт-Петербурге

```python
# cоздадим датафрейм, содержащий данные о продажах квартир в Санкт-Петербурге и расстояние которых до центра больше 0
saint_pet = data[(data['locality_name'] == 'Санкт-Петербург') & (data['cityCenters_nearest_km'] > 0)] 
```

```python
# проверим остутсвие пропусков
saint_pet.isna().sum() 
```

```python
# созадим новый столбец со стоимостью километра до центра
saint_pet['price_cityCenter_km'] = saint_pet['last_price'] / saint_pet['cityCenters_nearest_km'] 
```

```python
# построим график зависимости стоимости квартиры от расстояния до центра
sns.lineplot(data=saint_pet, 
             x='cityCenters_nearest_km', 
             y='last_price') 

plt.title('Зависимость стоимости квартиры от расстояния до центра')
plt.xlabel('Растояние до центра, км')
plt.ylabel('Стоимость квартиры, тыс.руб.');
```

Из данного графика видно, что чем ближе квартира расположена к центру, тем больше ее стоимость. Обратно пропорциональная связь.
